Each row represents a LHC collision event. Each column contains some observable from that event. The variable names are ([based on documentation](https://archive.ics.uci.edu/ml/datasets/SUSY)):

```python
VarNames=["signal", "l_1_pT", "l_1_eta","l_1_phi", "l_2_pT", "l_2_eta", "l_2_phi", "MET", "MET_phi", "MET_rel", "axial_MET", "M_R", "M_TR_2", "R", "MT2", "S_R", "M_Delta_R", "dPhi_r_b", "cos_theta_r1"]
```

Some of these variables represent the "raw" kinematics of the observed final state particles, while others are "features" that are derived from these raw quantities:

```python
RawNames=["l_1_pT", "l_1_eta","l_1_phi", "l_2_pT", "l_2_eta", "l_2_phi", "MET", "MET_phi"]
FeatureNames=list(set(VarNames[1:]).difference(RawNames))
```

```python
# defining low and high-level features
low_level = ["l_1_pT", "l_1_eta", "l_1_phi", "l_2_pT", "l_2_eta"]
high_level = ["M_TR_2", "MT2", "M_Delta_R", "dPhi_r_b", "cos_theta_r1"]

# creating pair plots
def pair_plot(df, features, sample_size=5000):
    df_sampled = df.sample(n=sample_size, random_state=42) 
    n_columns = len(features) #determine grid size based in number of features
    plt.figure(figsize=(15, 15)) # a figure to fit all subplot
    plot_i = 0  # keep track of subplot position
    for i, x_var_name in enumerate(features):
        for j, y_var_name in enumerate(features): # loop for column and row
            plot_i += 1 # subplot increment index
            plt.subplot(n_columns, n_columns, plot_i) # place each plot in a grid
            make_legend = plot_i == 1  # legend for first plot
            if i == j:
                # for diagonal histograms
                plt.hist(df_sampled[x_var_name], bins=50, alpha=0.5, density=True, color="blue")
                plt.ylabel("Density") # label
            else:
                # for scatter and 2d histograms
                plt.hist2d(df_sampled[y_var_name], df_sampled[x_var_name], bins=50, cmap="Blues")
            # labels
            if i == n_columns - 1:
                plt.xlabel(y_var_name)
            if j == 0:
                plt.ylabel(x_var_name)
    plt.tight_layout()
    plt.show()
# run 
pair_plot(df, low_level)
pair_plot(df, high_level)
```

```python
#Part b:
import numpy as np

# compute covariance and correlation for all features
cov_all = np.cov(df, rowvar=False)
corr_all = np.corrcoef(df, rowvar=False)

# extracting low and high-level features columns
low_level_data = df[low_level].values
high_level_data = df[high_level].values

# covariance and correlation matrices for low-level features
cov_low = np.cov(low_level_data, rowvar=False)
corr_low = np.corrcoef(low_level_data, rowvar=False)

# covariance and correlation matrices for high-level features
cov_high = np.cov(high_level_data, rowvar=False)
corr_high = np.corrcoef(high_level_data, rowvar=False)
```

```python
#Part c:
from tabulate import tabulate
from IPython.display import HTML, display

# function to format and display covariance and correlation matrices
def display_matrices(cov_matrix, corr_matrix, feature_names, matrix_type="Covariance"):
    # Format matrix headers
    headers = feature_names
    
    # tabulate both covariance and correlation matrices
    cov_table = [[f"{cov_matrix[i][j]:.3f}" for j in range(len(cov_matrix))] for i in range(len(cov_matrix))]
    corr_table = [[f"{corr_matrix[i][j]:.3f}" for j in range(len(corr_matrix))] for i in range(len(corr_matrix))]
    # create the formatted tables for display
    display(HTML(tabulate(cov_table, headers, tablefmt="html", numalign="right", showindex=True)))
    display(HTML(tabulate(corr_table, headers, tablefmt="html", numalign="right", showindex=True)))
    
# display covariance and correlation for all features
print("Covariance Matrix for All Features")
display_matrices(cov_all, corr_all, df.columns)

# display covariance and correlation for low-level features
print("Covariance Matrix for Low-Level Features")
display_matrices(cov_low, corr_low, low_level)

# display covariance and correlation for high-level features
print("Covariance Matrix for High-Level Features")
display_matrices(cov_high, corr_high, high_level)
```

```python
#Part d:
def compute_and_display_matrices(df, low_level, high_level):
    # compute covariance and correlation matrices for all features
    cov_all = np.cov(df, rowvar=False)
    corr_all = np.corrcoef(df, rowvar=False)

    # covariance and correlation for low-level features
    low_level_data = df[low_level].values
    cov_low = np.cov(low_level_data, rowvar=False)
    corr_low = np.corrcoef(low_level_data, rowvar=False)
    # covariance and correlation for high-level features
    high_level_data = df[high_level].values
    cov_high = np.cov(high_level_data, rowvar=False)
    corr_high = np.corrcoef(high_level_data, rowvar=False)

    # display matrices using tabulate
    print("Covariance and Correlation Matrices for All Features")
    display_matrices(cov_all, corr_all, df.columns)
    print("Covariance and Correlation Matrices for Low-Level Features")
    display_matrices(cov_low, corr_low, low_level)
    print("Covariance and Correlation Matrices for High-Level Features")
    display_matrices(cov_high, corr_high, high_level)
# run 
compute_and_display_matrices(df, low_level, high_level)
```
