2. Add the following methods:
    * `shape()`: returns a tuple `(n,m)` of the shape of the matrix.
    * `transpose()`: returns a new matrix instance which is the transpose of the matrix.
    * `row(n)` and `column(n)`: that return the nth row or column of the matrix M as a new appropriately shaped matrix object.
    * `to_list()`: which returns the matrix as a list of lists.
    *  `block(n_0,n_1,m_0,m_1)` that returns a smaller matrix located at the n_0 to n_1 columns and m_0 to m_1 rows. 
    * (Extra credit) Modify `__getitem__` implemented above to support slicing.
        

```python
class Matrix:
    def __init__(self, *args):
        if len(args) == 2:  # create matrix with n (rows) and m (columns) size
            n, m = args
            self.data = [[0] * m for _ in range(n)]
        elif len(args) == 1 and isinstance(args[0], list):  # create matrix with a list of lists
            self.data = args[0]
            # verify that all rows have the same number of columns
            row_lengths = [len(row) for row in self.data]
            if len(set(row_lengths)) > 1:
                raise ValueError("All rows must have the same number of columns.")
        else:
            raise ValueError("Invalid arguments. Provide either two integers (n, m) or a list of lists.")
    
    def __getitem__(self, index):
        if isinstance(index, tuple):
            i, j = index
            return self.data[i][j]
        elif isinstance(index, list):
            return self.data[index[0]]
        elif isinstance(index, slice):  # Supports slicing
            start, stop, step = index.indices(len(self.data))
            return [row[start:stop:step] for row in self.data]
        else:
            raise TypeError("Index must be a tuple (i, j), list [i], or slice.")
    
    def __setitem__(self, index, value):
        if isinstance(index, tuple):
            i, j = index
            self.data[i][j] = value
        else:
            raise TypeError("Index must be a tuple (i, j).")
    
    def __repr__(self):
        return '\n'.join([str(row) for row in self.data])

    # method to return the shape of the matrix
    def shape(self):
        return (len(self.data), len(self.data[0]) if self.data else 0)

    # method to return the transpose of the matrix
    def transpose(self):
        return Matrix([[self.data[j][i] for j in range(len(self.data))] for i in range(len(self.data[0]))])

    # method to return the nth row as a new matrix
    def row(self, n):
        if 0 <= n < len(self.data):
            return Matrix([self.data[n]])
        else:
            raise IndexError("Row index out of range.")

    # method to return the nth column as a new matrix
    def column(self, n):
        if len(self.data) == 0 or not (0 <= n < len(self.data[0])):
            raise IndexError("Column index out of range.")
        return Matrix([[self.data[i][n]] for i in range(len(self.data))])

    # method to return the matrix as a list of lists
    def to_list(self):
        return [row[:] for row in self.data]

    # method to return a block of the matrix defined by n_0, n_1, m_0, m_1
    def block(self, n_0, n_1, m_0, m_1):
        if not (0 <= n_0 < n_1 <= len(self.data)) or not (0 <= m_0 < m_1 <= len(self.data[0])):
            raise IndexError("Block indices out of range.")
        return Matrix([row[m_0:m_1] for row in self.data[n_0:n_1]])

# example usage:
# create a matrix
m = Matrix([[1, 2, 3], [4, 5, 6], [7, 8, 9]])

# Shape
print("Shape of m:", m.shape())

# Transpose
print("\nTranspose of m:\n", m.transpose())

# Row and column
print("\nRow 1 of m:\n", m.row(1))
print("\nColumn 2 of m:\n", m.column(2))

# Convert to list
print("\nMatrix as a list of lists:", m.to_list())

# Block from row 0 to 2, column 0 to 2
print("\nBlock from (0, 2) and (0, 2):\n", m.block(0, 2, 0, 2))

# Slicing
print("\nSliced matrix (rows 0 to 2, columns 0 to 2):")
print(m[0:2, 0:2])
```

Write functions that create special matrices (note these are standalone functions, not member functions of your matrix class):
constant(n,m,c): returns a n by m matrix filled with floats of value c.
zeros(n,m) and ones(n,m): return n by m matrices filled with floats of value 0 and 1, respectively.
eye(n): returns the n by n identity matrix

```python
def constant(n, m, c):
    # return an n x m matrix filled with float(c)
    return [[float(c) for _ in range(m)] for _ in range(n)]

def zeros(n, m):
   # return an n x m matrix filled with 0.0
    return constant(n, m, 0)

def ones(n, m):
   # return an n x m matrix filled with 1.0
    return constant(n, m, 1)

def eye(n):
    # return an n x n identity matrix (floats)
    return [[1.0 if i == j else 0.0 for j in range(n)] for i in range(n)]
```

```python
# create a 3x4 matrix with all values set to 7.
const_matrix = constant(3, 4, 7)
print("3x4 Constant Matrix with value 7:")
for row in const_matrix:
    print(row)

# create a 2x5 matrix of zeros.
zero_matrix = zeros(2, 5)
print("\n2x5 Zero Matrix:")
for row in zero_matrix:
    print(row)

# create a 4x3 matrix of ones.
one_matrix = ones(4, 3)
print("\n4x3 One Matrix:")
for row in one_matrix:
    print(row)

# create a 5x5 identity matrix.
identity_matrix = eye(5)
print("\n5x5 Identity Matrix:")
for row in identity_matrix:
    print(row)
```

4. Add the following member functions to your class. Make sure to appropriately test the dimensions of the matrices to make sure the operations are correct.
    * `M.scalarmul(c)`: a matrix that is scalar product $cM$, where every element of $M$ is multiplied by $c$.
    * `M.add(N)`: adds two matrices $M$ and $N$. Don’t forget to test that the sizes of the matrices are compatible for this and all other operations.
    * `M.sub(N)`: subtracts two matrices $M$ and $N$.
    * `M.mat_mult(N)`: returns a matrix that is the matrix product of two matrices $M$ and $N$.
    * `M.element_mult(N)`: returns a matrix that is the element-wise product of two matrices $M$ and $N$.
    * `M.equals(N)`: returns true/false if $M==N$.

```python
class Matrix:
    def __init__(self, *args):
        if len(args) == 2:  # create matrix with n (rows) and m (columns) size
            n, m = args
            self.data = [[0] * m for _ in range(n)]
        elif len(args) == 1 and isinstance(args[0], list):  # create matrix with a list of lists
            self.data = args[0]
            # verify that all rows have the same number of columns
            row_lengths = [len(row) for row in self.data]
            if len(set(row_lengths)) > 1:
                raise ValueError("All rows must have the same number of columns.")
        else:
            raise ValueError("Invalid arguments. Provide either two integers (n, m) or a list of lists.")
    
    def __getitem__(self, index):
        if isinstance(index, tuple):
            i, j = index
            return self.data[i][j]
        elif isinstance(index, list):
            return self.data[index[0]]
        elif isinstance(index, slice):  # supports slicing
            start, stop, step = index.indices(len(self.data))
            return [row[start:stop:step] for row in self.data]
        else:
            raise TypeError("Index must be a tuple (i, j), list [i], or slice.")
    
    def __setitem__(self, index, value):
        if isinstance(index, tuple):
            i, j = index
            self.data[i][j] = value
        else:
            raise TypeError("Index must be a tuple (i, j).")
    
    def __repr__(self):
        return '\n'.join([str(row) for row in self.data])

    # method to return the shape of the matrix
    def shape(self):
        return (len(self.data), len(self.data[0]) if self.data else 0)

    # method to return the transpose of the matrix
    def transpose(self):
        return Matrix([[self.data[j][i] for j in range(len(self.data))] for i in range(len(self.data[0]))])

    # method to return the nth row as a new matrix
    def row(self, n):
        if 0 <= n < len(self.data):
            return Matrix([self.data[n]])
        else:
            raise IndexError("Row index out of range.")

    # method to return the nth column as a new matrix
    def column(self, n):
        if len(self.data) == 0 or not (0 <= n < len(self.data[0])):
            raise IndexError("Column index out of range.")
        return Matrix([[self.data[i][n]] for i in range(len(self.data))])

    # method to return the matrix as a list of lists
    def to_list(self):
        return [row[:] for row in self.data]

    # method to return a block of the matrix defined by n_0, n_1, m_0, m_1
    def block(self, n_0, n_1, m_0, m_1):
        if not (0 <= n_0 < n_1 <= len(self.data)) or not (0 <= m_0 < m_1 <= len(self.data[0])):
            raise IndexError("Block indices out of range.")
        return Matrix([row[m_0:m_1] for row in self.data[n_0:n_1]])

    # scalar multiplication (𝑐M)
    def scalarmul(self, c):
        return Matrix([[c * self.data[i][j] for j in range(len(self.data[i]))] for i in range(len(self.data))])

    # matrix addition (M + N)
    def add(self, N):
        if self.shape() != N.shape():
            raise ValueError("Matrices must have the same shape to be added.")
        return Matrix([[self.data[i][j] + N[i, j] for j in range(len(self.data[i]))] for i in range(len(self.data))])

    # matrix subtraction (M - N)
    def sub(self, N):
        if self.shape() != N.shape():
            raise ValueError("Matrices must have the same shape to be subtracted.")
        return Matrix([[self.data[i][j] - N[i, j] for j in range(len(self.data[i]))] for i in range(len(self.data))])

    # matrix multiplication (M * N)
    def mat_mult(self, N):
        if len(self.data[0]) != len(N.data):
            raise ValueError("Number of columns in M must be equal to number of rows in N for multiplication.")
        return Matrix([[sum(self.data[i][k] * N[k, j] for k in range(len(self.data[0]))) for j in range(len(N.data[0]))] for i in range(len(self.data))])

    # element-wise multiplication (M .* N)
    def element_mult(self, N):
        if self.shape() != N.shape():
            raise ValueError("Matrices must have the same shape for element-wise multiplication.")
        return Matrix([[self.data[i][j] * N[i, j] for j in range(len(self.data[i]))] for i in range(len(self.data))])

    # equality check (M == N)
    def equals(self, N):
        if self.shape() != N.shape():
            return False
        for i in range(len(self.data)):
            for j in range(len(self.data[i])):
                if self.data[i][j] != N[i, j]:
                    return False
        return True

# example usage:

# Matrix initialization
M = Matrix([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
N = Matrix([[9, 8, 7], [6, 5, 4], [3, 2, 1]])

# Scalar multiplication
print("Scalar multiplication of M by 2:\n", M.scalarmul(2))

# Matrix addition
print("\nM + N:\n", M.add(N))

# Matrix subtraction
print("\nM - N:\n", M.sub(N))

# Matrix multiplication
print("\nMatrix multiplication M * N:\n", M.mat_mult(N))

# Element-wise multiplication
print("\nElement-wise multiplication M .* N:\n", M.element_mult(N))

# Matrix equality
print("\nM equals N?", M.equals(N))

```

5. Overload python operators to appropriately use your functions in 4 and allow expressions like:
    * 2*M
    * M*2
    * M+N
    * M-N
    * M*N
    * M==N
    * M=N


```python
class Matrix:
    def __init__(self, data, m=None):
        if isinstance(data, list):  # Initialize with list of lists
            if not all(isinstance(row, list) for row in data):
                raise ValueError("Matrix must be a list of lists.")
            row_lengths = set(len(row) for row in data)
            if len(row_lengths) > 1:
                raise ValueError("All rows must have the same number of columns.")
            self.data = data
            self.n = len(data)
            self.m = len(data[0])
        elif isinstance(data, int) and isinstance(m, int):  # Initialize with dimensions
            self.n, self.m = data, m
            self.data = [[0] * m for _ in range(data)]
        else:
            raise ValueError("Invalid initialization parameters.")

    def shape(self):
        return self.n, self.m

    def transpose(self):
        transposed_data = [[self.data[j][i] for j in range(self.n)] for i in range(self.m)]
        return Matrix(transposed_data)

    def row(self, n):
        return Matrix([self.data[n]])

    def column(self, n):
        return Matrix([[self.data[i][n]] for i in range(self.n)])

    def to_list(self):
        return [row[:] for row in self.data]

    def block(self, n_0, n_1, m_0, m_1):
        return Matrix([row[m_0:m_1] for row in self.data[n_0:n_1]])

    def __getitem__(self, index):
        if isinstance(index, tuple):
            i, j = index
            if isinstance(i, slice) or isinstance(j, slice):
                return Matrix([row[j] for row in self.data[i]])
            return self.data[i][j]
        return self.data[index]

    def __setitem__(self, index, value):
        if isinstance(index, tuple) and len(index) == 2:
            i, j = index
            self.data[i][j] = value
        else:
            raise IndexError("Invalid index format. Use M[i, j].")

    def assign(self, other):
        if isinstance(other, Matrix):
            if self.n == other.n and self.m == other.m:
                self.data = [row[:] for row in other.data]
            else:
                raise ValueError("Matrix dimensions must match for assignment.")
        elif isinstance(other, list) and len(other) == self.n and all(len(row) == self.m for row in other):
            self.data = [row[:] for row in other]
        else:
            raise ValueError("Invalid assignment value.")

    def __repr__(self):
        return "\n".join(str(row) for row in self.data)

    # member functions for operations

    def scalarmul(self, c):
        result = [[self.data[i][j] * c for j in range(self.m)] for i in range(self.n)]
        return Matrix(result)

    def add(self, other):
        if not isinstance(other, Matrix):
            raise ValueError("Operand must be a Matrix.")
        if self.n != other.n or self.m != other.m:
            raise ValueError("Matrices must have the same dimensions for addition.")
        result = [
            [self.data[i][j] + other.data[i][j] for j in range(self.m)]
            for i in range(self.n)
        ]
        return Matrix(result)

    def sub(self, other):
        if not isinstance(other, Matrix):
            raise ValueError("Operand must be a Matrix.")
        if self.n != other.n or self.m != other.m:
            raise ValueError("Matrices must have the same dimensions for subtraction.")
        result = [
            [self.data[i][j] - other.data[i][j] for j in range(self.m)]
            for i in range(self.n)
        ]
        return Matrix(result)

    def mat_mult(self, other):
        if not isinstance(other, Matrix):
            raise ValueError("Operand must be a Matrix.")
        if self.m != other.n:
            raise ValueError("Matrix multiplication not possible: self.m must equal other.n.")
        result = [
            [sum(self.data[i][k] * other.data[k][j] for k in range(self.m))
             for j in range(other.m)]
            for i in range(self.n)
        ]
        return Matrix(result)

    def element_mult(self, other):
        if not isinstance(other, Matrix):
            raise ValueError("Operand must be a Matrix.")
        if self.n != other.n or self.m != other.m:
            raise ValueError("Matrices must have the same dimensions for element-wise multiplication.")
        result = [
            [self.data[i][j] * other.data[i][j] for j in range(self.m)]
            for i in range(self.n)
        ]
        return Matrix(result)

    def equals(self, other):
        if not isinstance(other, Matrix):
            return False
        if self.n != other.n or self.m != other.m:
            return False
        for i in range(self.n):
            for j in range(self.m):
                if self.data[i][j] != other.data[i][j]:
                    return False
        return True

    # operator overloading to support 2*M, M*2, M+N, M-N, M*N, M==N

    def __add__(self, other):
        if isinstance(other, Matrix):
            return self.add(other)
        return NotImplemented

    def __sub__(self, other):
        if isinstance(other, Matrix):
            return self.sub(other)
        return NotImplemented

    def __mul__(self, other):
        if isinstance(other, (int, float)):
            return self.scalarmul(other)
        elif isinstance(other, Matrix):
            return self.mat_mult(other)
        return NotImplemented

    def __rmul__(self, other):
        if isinstance(other, (int, float)):
            return self.scalarmul(other)
        return NotImplemented

    def __eq__(self, other):
        return self.equals(other)


# test usage examples for overloaded operators
if __name__ == "__main__":
    # Create two 2x2 matrices
    M = Matrix([[1, 2], [3, 4]])
    N = Matrix([[5, 6], [7, 8]])

    # Scalar multiplication
    print("2 * M:")
    print(2 * M)
    print("\nM * 2:")
    print(M * 2)

    # Matrix addition
    print("\nM + N:")
    print(M + N)

    # Matrix subtraction
    print("\nM - N:")
    print(M - N)

    # Matrix multiplication (M * N)
    print("\nM * N (matrix multiplication):")
    print(M * N)

    # Equality test
    print("\nM == Matrix([[1, 2], [3, 4]]):", M == Matrix([[1, 2], [3, 4]]))
    print("M == N:", M == N)
```
